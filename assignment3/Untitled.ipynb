```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

def mapping(points):
#     points = points / np.max(points, axis=1,keepdims=True)
    base_x = np.array([0, 1])
    base_y = np.array([3**0.5 / 2, -0.5])
    base_z = np.array([-3**0.5 / 2, -0.5])
    M = np.array([base_x, base_y, base_z])
    return points.dot(M)


color_list = []
x_list = []
y_list = []
z_list = []
X = np.arange(0,1,0.05)
N= 40000
for n in range(N):
    z = (2 * n + 1) / N - 1
    x = np.sqrt(1 - z**2) * np.cos(np.pi*n*(5**0.5-1))
    y = np.sqrt(1 - z**2) * np.sin(np.pi*n*(5**0.5-1))
    if(z > 0 and x > 0 and y > 0):
        scale = max(x,y,z)
        x_list.append(x/scale)
        y_list.append(y/scale)
        z_list.append(z/scale)
        color_list.append([x/scale,y/scale,z/scale])
points = np.concatenate([np.array([x_list]), np.array([y_list]), np.array([z_list])], axis = 0).T
new_points = mapping(points)
# print(points)

# for phi in np.arange(0,np.pi/2,0.05*np.pi/2):
#     for theta in np.arange(0,np.pi/2,0.05*np.pi/2):
#         x = np.sin(theta) * np.cos(phi)
#         y = np.sin(theta) * np.sin(phi)
#         z = np.cos(theta)
#         scale = max(x,y,z)
#         x_list.append(x)
#         y_list.append(y)
#         z_list.append(z)
#         color_list.append([x/scale,y/scale,z/scale])
# for x in X:
#     for y in np.arange(0, np.sqrt(1-x**2),0.01):
#         z = np.sqrt(1-x**2-y**2)
        
# y = np.arange(0,1,0.01)
# X, Y = np.meshgrid(x,y)
# Z = np.sqrt(np.maximum(0, 1- X**2 - Y**2))
# # c = np.random.standard_normal(100)
# color = np.concatenate([X[:,:,None],Y[:,:,None],Z[:,:,None]],axis = 2)
# print(color.shape)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_list, y_list, z_list, c=color_list)
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
# print(new_points.shape)
ax2.scatter(new_points[:,0], new_points[:,1], c=color_list)
# ax.scatter(X, Y, Z, color = list(color))
plt.show()
```
